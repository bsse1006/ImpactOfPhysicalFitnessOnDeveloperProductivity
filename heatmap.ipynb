```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
dataset = pd.read_csv("EncodedData.xlsx")
dataset.dropna(inplace=True)
dataset.drop(columns=['Unnamed: 0'],inplace=True)
print(dataset.shape)
list(dataset.columns.values)
```

```python
data=dataset[['Decision Contribution','Spontaneous Participation','Task Satisfaction','Interaction Willingness',
             'Average Sleep Duration','Average Sleep Starting Time','Average Breakfast Time','Average Lunch Time','Average Dinner Time']]
cor=data.corr(method='spearman')
# setting the parameter values
annot = True
  
# plotting the heatmap
hm = sn.heatmap(data=cor,
                annot=annot)
  
# displaying the plotted heatmap
plt.show()
```
